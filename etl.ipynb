```python
import pandas as pd
import numpy as np
# from sqlalchemy import create_engine

pd.options.display.max_columns = None

# import 
chef_payment=pd.read_json("chefmozaccepts.json")
chef_cuisine=pd.read_json("chefmozcuisine.json")
chef_hours=pd.read_json("chefmozhours4_final_final_final.json")
chef_parking=pd.read_json("chefmozparking.json")

restaurent_merged_table_1 = chef_payment.merge(chef_cuisine, on='placeID', how='inner')
restaurent_merged_table_2 = restaurent_merged_table_1.merge(chef_hours, on='placeID', how='inner')
restaurent_table = restaurent_merged_table_2.merge(chef_parking, on='placeID', how='inner')

user_cuisine=pd.read_excel("usercuisine.xlsx")
user_profile=pd.read_excel("userprofile.xlsx")
user_payment=pd.read_excel("userpayment.xlsx")

user_merged_table_2 = user_profile.merge(user_cuisine, on='userID', how='inner')
user_table=user_merged_table_2.merge(user_payment, on='userID', how='inner')

rating = pd.read_csv('rating_final.txt', delimiter=' ')

display(user_table.head(20))
user_table.describe()
user_table.info()
display(restaurent_table.head(20))
restaurent_table.describe()
restaurent_table.info()
display(rating)
rating.describe()
rating.info()
# engine = create_engine("postgresql://username:password@localhost:5432/dbms")

# restaurent_table.to_sql("restaurent", engine)
# user_table.to_sql("user",engine)
```

```python
user_table.replace("?",np.nan)
result_user=user_table.groupby('userID', as_index=False).agg({
    'latitude': 'mean',
    'longitude': 'mean',
    'drink_level': 'first',
    'dress_preference': 'first',
    'ambience': 'first',
    'transport': 'first',
    'marital_status': 'first',
    'hijos': 'first',
    'birth_year': 'mean',
    'interest': 'first',
    'personality': 'first',
    'religion': 'first',
    'activity': 'first',
    'color': 'first',
    'weight': 'mean',
    'budget': 'first',
    'height': 'mean',
    'Rcuisine': lambda x: list(set(x)),
    'Upayment': lambda x: list(set(x))
}).assign(birth_year = lambda x: x.birth_year.astype(int))
# Specific values to remove
outliers = ['Australian', 'Austrian', 'Basque', 'British', 'Burmese', 'Cajun-Creole', 'Canadian', 'Chilean', 'Cuban', 'Dim_Sum', 'Doughnuts', 'Eclectic', 'Filipino', 'Fusion', 'Hawaiian', 'Hungarian', 'Indian-Pakistani', 'Indigenous', 'Indonesian', 'Irish', 'Israeli', 'Jamaican', 'Kosher', 'Lebanese', 'Malaysian', 'Moroccan', 'North_African', 'Pacific_Northwest', 'Pacific_Rim', 'Peruvian', 'Polynesian', 'Portuguese', 'Romanian', 'Russian-Ukrainian', 'Scandinavian', 'Southeast_Asian', 'Swiss', 'Tapas', 'Tea_House', 'Tibetan', 'Tunisian','Middle_Eastern','Moroccan','Organic-Healthy','Persian','Peruvian','Polish','Polynesian','Portuguese','Regional','Romanian','Russian-Ukrainian','Scandinavian','Seafood','Soup','Southeast_Asian','Southern','Southwestern','Spanish','Steaks','Sushi','Swiss','Tapas','Tea_House','Tex-Mex','Thai','Tibetan','Tunisian','Turkish','Vegetarian','Vietnamese']

# Iterating through the rows of the dataframe
for index, row in result_user.iterrows():
    for outlier in outliers:
        if outlier in row['Rcuisine']:
            row['Rcuisine'].remove(outlier)
result_user_cuisine = result_user[['userID','Rcuisine']].copy()
result_user = result_user.drop('Rcuisine', axis=1)
result_user_cuisine = result_user_cuisine.explode('Rcuisine')
result_user_payment = result_user[['userID','Upayment']].copy()
result_user = result_user.drop('Upayment', axis=1)
result_user_payment = result_user_payment.explode('Upayment')
result_user_cuisine = result_user_cuisine.dropna(axis=0)
display(result_user)
display(result_user_cuisine)
display(result_user_payment)


```

```python

restaurent_table.replace("?",np.nan).head(20)
result_restaurent = restaurent_table.groupby('placeID', as_index=False).agg({
    'Rpayment': lambda x: list(set(x)),
    'Rcuisine': lambda x: list(set(x)),
    'hours': 'first',
    'days': 'first',
    'parking_lot': 'first'
})
result_restaurent_cuisine = result_restaurent[['placeID','Rcuisine']].copy()
result_restaurent = result_restaurent.drop('Rcuisine', axis=1)
result_restaurent_cuisine = result_restaurent_cuisine.explode('Rcuisine')
result_restaurent_payment = result_restaurent[['placeID','Rpayment']].copy()
result_restaurent = result_restaurent.drop('Rpayment', axis=1)
result_restaurent_payment = result_restaurent_payment.explode('Rpayment')
result_restaurent_work_hours = result_restaurent[['placeID','hours','days']].copy()
result_restaurent = result_restaurent.drop(['hours','days'], axis=1)
result_restaurent_work_hours = result_restaurent_work_hours.explode(['hours','days'])
display(result_restaurent_cuisine)
display(result_restaurent_payment)
display(result_restaurent_work_hours)
display(result_restaurent)
```

```python
valid_user_ids = set(result_user['userID'])
valid_place_ids = set(result_restaurent['placeID'])
valid_rows = rating.apply(lambda x: x['userID'] in valid_user_ids and x['placeID'] in valid_place_ids, axis=1)
display(valid_rows)

filtered_result_df = rating[valid_rows]
display(filtered_result_df)
```

```python
import os
if not os.path.exists("results"):
    os.makedirs("results")
    
result_user.to_json('results/users.json',orient='records')
result_user_cuisine.to_json('results/users_cuisine.json',orient='records')
result_user_payment.to_json('results/users_payment.json',orient='records')
result_restaurent.to_json('results/restaurants.json',orient='records')
result_restaurent_cuisine.to_json('results/restaurants_cuisine.json',orient='records')
result_restaurent_payment.to_json('results/restaurants_payment.json',orient='records')
result_restaurent_work_hours.to_json('results/restaurants_work_hours.json',orient='records')


filtered_result_df.to_json('results/ratings.json',orient='records')
```
